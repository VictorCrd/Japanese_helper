# Merge csv 

```python
NLPT5 = pd.concat([Nfive_raw["Kana_Kanji"], passfive_raw["Kanji"], JLPT_N5["Japanese"]], ignore_index = True).drop_duplicates().reset_index(drop = True)
```

```python
NLPT4 = pd.concat([Nfor_raw["Kana_Kanji"], passfor_raw["Kanji"], JLPT_N4["Japanese"]], ignore_index = True).drop_duplicates().reset_index(drop = True)
```

```python
NLPT3 = pd.concat([passthree_raw["Kanji"], JLPT_N3["Japanese"]], ignore_index = True).drop_duplicates().reset_index(drop = True)
```

```python
NLPT5 = pd.DataFrame({'vocab':NLPT5})
NLPT4 = pd.DataFrame({'vocab':NLPT4})
NLPT3 = pd.DataFrame({'vocab':NLPT3})
```

```python
NLPT3["count"] = 0
NLPT4["count"] = 0
NLPT5["count"] = 0
```

# Find nbr of NLPT word by level

```python
def JLPT_lvl(df_subtitles):
    for i in range(0, len(NLPT3)):
        NLPT3["count"][i] = len(df_subtitles[df_subtitles['phrases'].str.contains(NLPT3["vocab"][i])])
        NLPT3_sum = NLPT3["count"].sum()
        
    for i in range(0, len(NLPT4)):
        NLPT4["count"][i] = len(df_subtitles[df_subtitles['phrases'].str.contains(NLPT4["vocab"][i])])
        NLPT4_sum = NLPT4["count"].sum()

    for i in range(0, len(NLPT5)):
        NLPT5["count"][i] = len(df_subtitles[df_subtitles['phrases'].str.contains(NLPT5["vocab"][i])])
        NLPT5_sum = NLPT5["count"].sum()
        
        return NLPT3_sum, NLPT4_sum, NLPT5_sum
```

```python
lemmas_5 = japanese_lemmas[0:500].copy()
lemmas_4 = japanese_lemmas[500:1000].copy().reset_index(drop = True)
lemmas_3 = japanese_lemmas[1000:2000].copy().reset_index(drop = True)
lemmas_2 = japanese_lemmas[2000:15000].copy().reset_index(drop = True)
```

```python
lemmas_5["count"] = 0
lemmas_4["count"] = 0
lemmas_3["count"] = 0
lemmas_2["count"] = 0
```

```python
def JLPT_lvl_lemmas(df_subtitles):
    for i in range(0, len(lemmas_5)):
        lemmas_5["count"][i] = len(df_subtitles[df_subtitles['phrases'].str.contains(lemmas_5["lemma"][i])])
        lemmas_5_sum = lemmas_5["count"].sum()
        
    for i in range(0, len(lemmas_4)):
        lemmas_4["count"][i] = len(df_subtitles[df_subtitles['phrases'].str.contains(lemmas_4["lemma"][i])])
        lemmas_4_sum = lemmas_4["count"].sum()

    for i in range(0, len(lemmas_3)):
        lemmas_3["count"][i] = len(df_subtitles[df_subtitles['phrases'].str.contains(lemmas_3["lemma"][i])])
        lemmas_3_sum = lemmas_3["count"].sum()
        
    for i in range(0, len(lemmas_2)):
        lemmas_2["count"][i] = len(df_subtitles[df_subtitles['phrases'].str.contains(lemmas_2["lemma"][i])])
        lemmas_2_sum = lemmas_2["count"].sum()
        
        return lemmas_5_sum, lemmas_4_sum, lemmas_3_sum, lemmas_2_sum
```
